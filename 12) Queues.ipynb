#### Queue using Lists

```python
class queueInArray:

    def __init__(self):
        self.__arr = []
        self.__count = 0
        self.__front = 0

    def enqueue(self, data):
        self.__arr.append(data)
        self.__count += 1

    def dequeue(self):
        if self.__count == 0:
            return -1
        ele = self.__arr[self.__front]
        self.__front += 1
        self.__count -= 1
        return ele

    def front(self):
        if self.__count == 0:
            return -1
        return self.__arr[self.__front]

    def size(self):
        return self.__count

    def isEmpty(self):
        return self.size() == 0


q = queueInArray()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
q.enqueue(4)

while q.isEmpty() is False:
    print(q.front())
    q.dequeue()

print(q.dequeue())
```

#### Queues using Linked List

```python

from sys import stdin


class Node:

    def __init__(self, element):
        self.element = element
        self.next = None


class Queue:

    def __init__(self):
        self.head, self.tail = None, None
        self.count = 0

    def getSize(self):
        return self.count

    def isEmpty(self):
        return self.getSize() == 0

    def enqueue(self, data):
        newNode = Node(data)

        if self.head is None:
            self.head = newNode
            self.tail = newNode
        else:
            self.tail.next = newNode
            self.tail = newNode
        self.count = self.count + 1

    def dequeue(self):
        if self.head is None:
            return -1
        else:
            temp = self.head.element
            self.head = self.head.next
            self.count = self.count - 1
        return temp

    def front(self):
        if self.head is None:
            return -1

        temp = self.head.element

        return temp



q = int(stdin.readline().strip())

queue = Queue()

while q > 0:

    inputs = stdin.readline().strip().split(" ")
    choice = int(inputs[0])

    if choice == 1:
        element = int(inputs[1])
        queue.enqueue(element)

    elif choice == 2:
        print(queue.dequeue())

    elif choice == 3:
        print(queue.front())

    elif choice == 4:
        print(queue.getSize())

    else:
        if queue.isEmpty():
            print("true")
        else:
            print("false")

    q -= 1
```

#### Queue using two stacks

```python
class QueueUsingTwoStacks:
    
    def __init__(self):
        self.s1 = []
        self.s2 = []
        
    def enqueue(self,data):
        while (len(self.s1) != 0):
            self.s2.append(self.s1.pop())
        self.s1.append(data)
        
        while (len(self.s2) != 0):
            self.s1.append(self.s2.pop())
        
    def dequeue(self):
        if (len(self.s1) == 0):
            return -1
        return self.s1.pop()
        
    def front(self):
        if (len(self.s1) == 0):
            return -1
        return self.s1[-1]
        
    def getSize(self):
        return len(self.s1)
        
    def isEmpty(self):
        return self.getSize() == 0
    
q = QueueUsingTwoStacks()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
q.enqueue(4)
q.enqueue(5)

while (q.isEmpty() is False):
    print(q.front())
    q.dequeue()
```

#### Stack Using 2 Queues

```python
import queue
from sys import stdin


class Stack:

    def __init__(self):
        self.q1 = queue.Queue()
        self.q2 = queue.Queue()

    def getSize(self):
        return self.q1.qsize()

    def isEmpty(self):
        return self.getSize() == 0

    def push(self, data):
        self.q1.put(data)

    def pop(self):
        if self.isEmpty():
            return -1

        while self.q1.qsize() > 1:
            self.q2.put(self.q1.get())

        ans = self.q1.get()

        temp = self.q1
        self.q1 = self.q2
        self.q2 = temp

        return ans

    def top(self):
        if self.isEmpty():
            return -1

        while self.q1.qsize() > 1:
            self.q2.put(self.q1.get())

        ans = self.q1.get()
        self.q2.put(ans)

        temp = self.q1
        self.q1 = self.q2
        self.q2 = temp

        return ans


# main
q = int(stdin.readline().strip())

stack = Stack()

while q > 0:

    inputs = stdin.readline().strip().split(" ")
    choice = int(inputs[0])

    if choice == 1:
        data = int(inputs[1])
        stack.push(data)

    elif choice == 2:
        print(stack.pop())

    elif choice == 3:
        print(stack.top())

    elif choice == 4:
        print(stack.getSize())

    else:
        if stack.isEmpty():
            print("true")
        else:
            print("false")

    q -= 1
```

#### Reverse Queue

```python

from sys import stdin, setrecursionlimit
import queue

setrecursionlimit(10 ** 6)


def reverseQueue(inputQueue):
    if inputQueue.empty():
        return

    item = inputQueue.queue[0]
    inputQueue.get()

    reverseQueue(inputQueue)

    inputQueue.put(item)


def takeInput():
    n = int(stdin.readline().strip())

    qu = queue.Queue()
    values = list(map(int, stdin.readline().strip().split()))

    for i in range(n):
        qu.put(values[i])

    return qu


# main
t = int(stdin.readline().strip())

while t > 0:

    qu = takeInput()
    reverseQueue(qu)

    while not qu.empty():
        print(qu.get(), end=" ")

    print()

    t -= 1
```

#### Reverse the First K Elements in the Queue

```python
from sys import setrecursionlimit
import queue


def reverseFirstK(q, k):
    if (q.empty() == True or k > q.qsize()):
        return
    if (k <= 0):
        return
    Stack = []
    for i in range(k):
        Stack.append(q.queue[0])
        q.get()
    while (len(Stack) != 0):
        q.put(Stack[-1])
        Stack.pop()
    for i in range(q.qsize() - k):
        q.put(q.queue[0])
        q.get()


setrecursionlimit(11000)
n, k = input().split()
n = int(n)
k = int(k)
li = [int(ele) for ele in input().split()]
q = queue.Queue()
for ele in li:
    q.put(ele)
reverseFirstK(q, k)
while (q.qsize() > 0):
    print(q.get(), end=' ')
    n -= 1
```
