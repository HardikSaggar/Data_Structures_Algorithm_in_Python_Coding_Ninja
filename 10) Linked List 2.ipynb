#### Delete every N Nodes

```python
from sys import stdin


class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


def skipMdeleteN(head, M, N):

    if M == 0 or head is None:
        return None

    if N == 0:
        return head

    currentNode = head
    temp = None

    while currentNode is not None:
        take = 0
        skip = 0

        while currentNode is not None and take < M:
            if temp is None:
                temp = currentNode
            else:
                temp.next = currentNode
                temp = currentNode

            currentNode = currentNode.next
            take += 1

        while currentNode is not None and skip < N:
            currentNode = currentNode.next
            skip += 1

    if temp is not None:
        temp.next = None

    return head


# Taking Input Using Fast I/O
def takeInput():
    head = None
    tail = None

    datas = list(map(int, stdin.readline().rstrip().split(" ")))

    i = 0
    while (i < len(datas)) and (datas[i] != -1):
        data = datas[i]
        newNode = Node(data)

        if head is None:
            head = newNode
            tail = newNode

        else:
            tail.next = newNode
            tail = newNode

        i += 1

    return head


def printLinkedList(head):

    while head is not None:
        print(head.data, end=" ")
        head = head.next

    print()


# main
t = int(stdin.readline().rstrip())

while t > 0:

    head = takeInput()
    m_n = stdin.readline().strip().split(" ")

    m = int(m_n[0])
    n = int(m_n[1])

    newHead = skipMdeleteN(head, m, n)
    printLinkedList(newHead)

    t -= 1
```

#### Swap two nodes of LL

```python
from sys import stdin

# Following is the Node class already written for the Linked List


class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


def swapNodes(head, i, j):
    # Your code goes here
    if i == j:
        return head
    temp = head
    prev = None
    c1 = p1 = None
    c2 = p2 = None
    count = 0
    while temp is not None:
        if count == i:
            c1 = temp
            p1 = prev
        elif count == j:
            c2 = temp
            p2 = prev
        prev = temp
        temp = temp.next
        count+=1

    if p1 is not None:
        p1.next = c2
    else:
        head = c2

    if p2 is not None:
        p2.next = c1
    else:
        head = c1

    p = c2.next
    c2.next = c1.next
    c1.next = p
    return head



def takeInput():
    head = None
    tail = None

    datas = list(map(int, stdin.readline().rstrip().split(" ")))

    i = 0
    while (i < len(datas)) and (datas[i] != -1):
        data = datas[i]
        newNode = Node(data)

        if head is None:
            head = newNode
            tail = newNode

        else:
            tail.next = newNode
            tail = newNode

        i += 1

    return head


def printLinkedList(head):

    while head is not None:
        print(head.data, end=" ")
        head = head.next

    print()


# main
t = int(stdin.readline().rstrip())

while t > 0:

    head = takeInput()
    i_j = stdin.readline().strip().split(" ")

    i = int(i_j[0])
    j = int(i_j[1])

    newHead = swapNodes(head, i, j)
    printLinkedList(newHead)

    t -= 1
```

#### Bubble Sort (Iterative) LinkedList

```python
from sys import stdin

# Following is the Node class already written for the Linked List


class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


def swapAlt(head, prev, first, sec):
    if prev is not None:
        prev.next = sec
    else:
        head = sec
    first.next = sec.next
    sec.next = first
    return head


def bubbleSort(head):
    length = 0
    temp = head
    while temp is not None:
        temp = temp.next
        length += 1
    k = 0
    ref = None
    while k <= length-2:
        temp = head
        prev = None
        while temp.next != ref:
            if temp.data > temp.next.data:
                if prev is None:
                    prev = temp.next
                    head = swapAlt(head, None, temp, temp.next)
                else:
                    nowPrev = prev
                    prev = temp.next
                    head = swapAlt(head, nowPrev, temp, temp.next)
            else:
                prev = temp
                temp = temp.next

        ref = temp
        k += 1
    return head


def takeInput():
    head = None
    tail = None

    datas = list(map(int, stdin.readline().rstrip().split(" ")))

    i = 0
    while (i < len(datas)) and (datas[i] != -1):
        data = datas[i]
        newNode = Node(data)

        if head is None:
            head = newNode
            tail = newNode

        else:
            tail.next = newNode
            tail = newNode

        i += 1

    return head


def printLinkedList(head):

    while head is not None:
        print(head.data, end=" ")
        head = head.next

    print()


# main
head = takeInput()
head = bubbleSort(head)
printLinkedList(head)
```
